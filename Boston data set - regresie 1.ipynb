```python
from sklearn.datasets import load_boston

load_boston()
```

```python
import pandas as pd

df = pd.DataFrame(load_boston().data, columns=load_boston().feature_names)

df
```

```python
df['MEDV'] = load_boston().target

df
```

```python
corr = df.corr()

import seaborn as sns

sns.heatmap(corr, annot=True)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=df['RM'], y=df['MEDV'], color='r')
plt.scatter(x=df['LSTAT'], y=df['MEDV'], color='b')


plt.show()
```

```python
coloane_x = ['RM','LSTAT', 'PTRATIO', 'TAX']
col_y = ['MEDV']

x = df[coloane_x]
y = df[col_y]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

print(X_train)
print(y_train)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X_train, y_train)

r_patrat = model.score(X_train, y_train)

print(r_patrat)

# f(x) = constanta + RM * x1 +LSTAT * x2 + PTRATIO * x3 + TAX*x4

print("constanta=", model.intercept_)
print ("b0=", model.coef_)
```
